```python
# MODLEL TRAINING
```

```python
X = df_cancer.drop(['target'], axis = 1)
```

```python
y = df_cancer['target']
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state=5)
```

```python
from sklearn.svm import SVC
```

```python
from sklearn.metrics import classification_report, confusion_matrix
```

```python
svc_model = SVC()
```

```python
svc_model.fit(X_train, y_train)
```

```python
# Evaluating the model
```

```python
y_predict = svc_model.predict(X_test)
```

```python
y_predict
```

```python
cm = confusion_matrix(y_test, y_predict)
```

```python
sns.heatmap(cm, annot = True)
```

```python
print(classification_report(y_test,y_predict))
```

```python
# Improving the Model
```

```python
# Unity Based optimization
```

```python
min_train = X_train.min()
```

```python
range_train = (X_train - min_train).max()
```

```python
X_train_scaled = (X_train - min_train)/range_train
```

```python
sns.scatterplot(x = X_train['mean area'], y = X_train['mean smoothness'], hue = y_train)
```

```python
sns.scatterplot(x = X_train_scaled['mean area'], y = X_train_scaled['mean smoothness'], hue = y_train)
```

```python
min_test = X_test.min()
range_test = (X_test - min_test).max()
X_test_scaled = (X_test - min_test)/range_test
```

```python
from sklearn.svm import SVC 
from sklearn.metrics import classification_report, confusion_matrix

svc_model = SVC()
svc_model.fit(X_train_scaled, y_train)
```

```python
y_predict = svc_model.predict(X_test_scaled)
cm = confusion_matrix(y_test, y_predict)

sns.heatmap(cm,annot=True,fmt="d")
```

```python
print(classification_report(y_test,y_predict))
```

```python
# SVM Parameters optimization
## C Parameters
### Small C(loose) - makes cost of miscalssification low (soft margin)
### Large C(strict) - makes cose fo miscalssification high (hard margin), forcing the model to explain input data
### stricter and potentially over fit
## Gamma Parameters
### Larget gamma - close reach (closer data points have high weight) 
### Small gamma - far reach
```

```python
param_grid = {'C': [0.1, 1, 10, 100], 'gamma': [1, 0.1, 0.01, 0.001], 'kernel': ['rbf']} 
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
grid = GridSearchCV(SVC(),param_grid,refit=True,verbose=4)
```

```python
grid.fit(X_train_scaled,y_train)
```

```python
grid.best_params_
```

```python
grid_predictions = grid.predict(X_test_scaled)
```

```python
cm = confusion_matrix(y_test, grid_predictions)
```

```python
sns.heatmap(cm, annot=True)
```
